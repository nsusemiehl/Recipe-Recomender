```python
import json
import pandas as pd
import re
```

```python
titles = []
ingredients = []
instructions = []
picture_links = []

with open(r"C:\Users\Nick\Desktop\data\recipes_raw_nosource_ar.json", "r") as read_file:
    ar_data = json.load(read_file)

for key, value in ar_data.items():
    if len(value) != 0:
        titles.append(value["title"])
        if value["ingredients"][-1] == "ADVERTISEMENT":
            value["ingredients"] = value["ingredients"][0:-1]
        ingredients.append(value["ingredients"])
        instructions.append(value["instructions"])
        picture_links.append(value["instructions"])
        
with open(r"C:\Users\Nick\Desktop\data\recipes_raw_nosource_epi.json", "r") as read_file:
    epi_data = json.load(read_file)

for key, value in epi_data.items():
    if len(value) != 0:
        titles.append(value["title"])
        ingredients.append(value["ingredients"])
        instructions.append(value["instructions"])
        picture_links.append(value["instructions"])
        
with open(r"C:\Users\Nick\Desktop\data\recipes_raw_nosource_fn.json", "r") as read_file:
    fn_data = json.load(read_file)

for key, value in fn_data.items():
    if len(value) != 0:
        titles.append(value["title"])
        ingredients.append(value["ingredients"])
        instructions.append(value["instructions"])
        picture_links.append(value["instructions"])
```

```python
er_df = pd.DataFrame({"Titles":titles, "Ingredients":ingredients, "Instructions":instructions, "Picture Links":picture_links})
display(er_df.head())
```

```python
er_df.to_csv(r"C:\Users\Nick\Desktop\data\er_dataframe.csv", index=False)
```
