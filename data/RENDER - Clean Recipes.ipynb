```python
nutritions_data = {'niacin': niacin_list, 'sugars': sugars_list, 'sodium': sodium_list, 
        'carbohydrates': carbohydrates_list, 'vitaminB6': vitaminB6_list, 'calories': calories_list,
        'thiamin': thiamin_list, 'fat': fat_list, 'folate': folate_list,
        'caloriesFromFat': caloriesFromFat_list, 'calcium': calcium_list, 'fiber': fiber_list, 
        'magnesium': magnesium_list, 'iron': iron_list, 'chalesterol': cholesterol_list,
        'protein': protein_list, 'vitaminA': vitaminA_list, 'potassium': potassium_list,
        'saturatedFat': saturatedFat_list, 'vitaminC': vitaminC_list}

nutritions_df = pd.DataFrame(nutritions_data)
nutritions_df.index = recipe['recipe_id']
nutritions_df.head()
```

```python
cooking_directions_dict = []

for row in recipe['cooking_directions']:
    cooking_directions_dict.append(ast.literal_eval(row))
    
cooking_directions_dict[0]
```

```python
list(cooking_directions_dict[0].values())[0]
list(cooking_directions_dict[0].values())[0].split('\n')
```

```python
prep_list = []
cook_list = []
ready_in_list = []
directions_list = []
directions_in_list_form = 0

for i in range(len(cooking_directions_dict)):
    temp = list(cooking_directions_dict[i].values())[0].split('\n')
    try:
        if temp[0] == 'Prep':
            if temp[2] == 'Cook':
                prep_list.append(temp[1])
                cook_list.append(temp[3])
                ready_in_list.append(temp[5])
                directions_list.append(' '.join(temp[6:]))
            elif temp[2] == 'Ready In':
                prep_list.append(temp[1])
                cook_list.append('NaN')
                ready_in_list.append(temp[3])
                directions_list.append(' '.join(temp[4:]))
        elif temp[0] == 'Cook':
            prep_list.append('NaN')
            if temp[2] == 'Ready In':
                cook_list.append(temp[1])
                ready_in_list.append(temp[3])
                directions_list.append(' '.join(temp[4:]))
            else:
                cook_list.append(temp[1])
                ready_in_list.append('NaN')
                directions_list.append(' '.join(temp[2:]))
        elif temp[0] == 'Ready In':
            prep_list.append('NaN')
            cook_list.append('NaN')
            ready_in_list.append(temp[1])
            directions_list.append(' '.join(temp[2:]))
        else:
            prep_list.append('NaN')
            cook_list.append('NaN')
            ready_in_list.append('NaN')
            directions_list.append(' '.join(temp))
            directions_in_list_form += 1
    except:
        prep_list.append('NaN')
        cook_list.append('NaN')
        ready_in_list.append('NaN')
        directions_list.append('NaN')

cooking_directions_data = {'prep': prep_list, 'cook': cook_list, 'ready_in': ready_in_list, 'directions': directions_list}

cooking_directions_df = pd.DataFrame(cooking_directions_data)
cooking_directions_df.index = recipe['recipe_id']

print(directions_in_list_form)
cooking_directions_df.head()
```

```python
null = cooking_directions_df['cook'] == 'NaN'
cooking_directions_df[null]
```

```python
# CODE TO FIND ready_in IN MINUTES
ready_in_min_list = []
temp = []
time =0
for row in range(len(cooking_directions_df.ready_in)):
    temp = ready_in_list[row].split(" ")
    for i in range(len(temp)):
        try:
            temp[i]= float(temp[i])
        except:
            pass
        if temp[i] == "h" or temp[i] == "H":
            try:
                temp[i-1] = 60*temp[i-1]
            except:
                pass
        time = sum([temp[i] for i in range(len(temp)) if type(temp[i]) == float ])
    ready_in_min_list.append(time)
    time = 0
    temp = []

ready_in_min_dict = {"ready_in_min": ready_in_min_list}

ready_in_min_df = pd.DataFrame(ready_in_min_dict)
ready_in_min_df.index = recipe.recipe_id
ready_in_min_list[0:100]
ready_in_min_df.head()
```

```python
# Separating the ingredients
ingredients_list = []
temp=[]
for row in range(len(recipe.ingredients)):
    temp = recipe.ingredients[row].split("^")
    ingredients_list.append(temp)
    temp = []

#  Cleaning the ingredients
ingredients_list_clean = []  
num_ingredients_list = []
for i in range(len(ingredients_list)):
    for j in range(len(ingredients_list[i])):
        if ":" not in ingredients_list[i][j]:
            temp.append(ingredients_list[i][j])
    ingredients_list_clean.append(temp)
    num_ingredients_list.append(len(temp))
    temp = []


ingredients_list_clean[0]
#num_ingredients_list[0]
```

```python
ingredients_dict = {"ingredients":ingredients_list_clean,"number_ingredients":num_ingredients_list}
ingredients_df = pd.DataFrame(ingredients_dict)
ingredients_df.index= recipe["recipe_id"]
ingredients_df.head()
```

```python
recipe_1 = recipe.drop(["ingredients","cooking_directions","image_url","cooking_directions","nutritions"],axis = 1)

recipe_clean = pd.merge(recipe_1, ingredients_df, on = "recipe_id")
recipe_clean = pd.merge(recipe_clean, cooking_directions_df, on = "recipe_id")
recipe_clean = pd.merge(recipe_clean, nutritions_df, on = "recipe_id")
recipe_clean = pd.merge(recipe_clean, ready_in_min_df, on = "recipe_id")

columns_names = list(recipe_clean.columns)
sorted_columns = ['recipe_id', 'recipe_name', 'ingredients', 'number_ingredients', 'prep', 'cook', 'ready_in',\
                  'ready_in_min', 'directions', 'niacin', 'sugars', 'sodium', 'carbohydrates', 'vitaminB6', 'calories',\
                  'thiamin', 'fat', 'folate', 'caloriesFromFat', 'calcium', 'fiber', 'magnesium', 'iron',\
                  'chalesterol', 'protein', 'vitaminA', 'potassium', 'saturatedFat', 'vitaminC' ]

recipe_clean = recipe_clean[sorted_columns]
#recipe_clean.head()
print(columns_names)
```

```python
from datetime import datetime
start = datetime.now()

ingredients_words_list= []
temp = []
for i in range(len(ingredients_list_clean)):
    for j in range(len(ingredients_list_clean[i])):
        temp = ingredients_list_clean[i][j].split(" ")
        for k in temp:
            k = k.replace("(", "")
            k = k.replace(")", "")
            k = k.replace("*", "")
            k = k.replace(u"\u00AE","") # to remove the "registered" sign
            
            ingredients_words_list.append(k.lower()) # all lowercase
        temp = []
        
ingredients_words_unique = list(set(ingredients_words_list))

ingredients_words_list_len =len(ingredients_words_list)
ingredients_words_unique_len = len(ingredients_words_unique)
#print(ingredients_words_list_len) ##1011708
#print(ingredients_words_unique) ##5519



count = 0
times_list= []
percent_list = []

for i in ingredients_words_unique:
    for j in ingredients_words_list:
        if i == j:
            count += 1
    times_list.append(count)
    percent_list.append(count*100/ingredients_words_list_len)
    count = 0
    
ingredients_words_dict = {"word":ingredients_words_unique, "times":times_list , "percent":percent_list}
    
ingredients_words_df = pd.DataFrame(ingredients_words_dict)

ingredients_words_df.sort_values(by=["percent"], ascending = False, inplace = True)


print (datetime.now()- start)##0:09:27.019229


ingredients_words_df.to_csv("ingredients_words.csv")
ingredients_words_df.head()
```
