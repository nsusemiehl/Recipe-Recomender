Kaggle (foodRecSys-V1; https://www.kaggle.com/elisaxxygao/foodrecsysv1)

```python
recipe = pd.read_csv(r'C:\Users\Nick\Desktop\data\core-data_recipe.csv')
recipe.head()

nutritions_dict = []
for row in recipe.nutritions:
    nutritions_dict.append(ast.literal_eval(row))
    
nutritions_keys = []
for i in range(len(nutritions_dict)):
    for key in list(nutritions_dict[i].keys()):
        if key in nutritions_keys:
            pass
        else:
            nutritions_keys.append(key)
            
niacin_list = []
sugars_list = []
sodium_list = []
carbohydrates_list = []
vitaminB6_list = []
calories_list = []
thiamin_list = []
fat_list = []
folate_list = []
caloriesFromFat_list = []
calcium_list = []
fiber_list = []
magnesium_list = []
iron_list = []
cholesterol_list = []
protein_list = []
vitaminA_list = []
potassium_list = []
saturatedFat_list = []
vitaminC_list = []

for i in range(len(nutritions_dict)):
    niacin_list.append(nutritions_dict[i]['niacin']['percentDailyValue'])
    sugars_list.append(nutritions_dict[i]['sugars']['percentDailyValue'])
    sodium_list.append(nutritions_dict[i]['sodium']['percentDailyValue'])
    carbohydrates_list.append(nutritions_dict[i]['carbohydrates']['percentDailyValue'])
    vitaminB6_list.append(nutritions_dict[i]['vitaminB6']['percentDailyValue'])
    calories_list.append(nutritions_dict[i]['calories']['percentDailyValue'])
    thiamin_list.append(nutritions_dict[i]['thiamin']['percentDailyValue'])
    fat_list.append(nutritions_dict[i]['fat']['percentDailyValue'])
    folate_list.append(nutritions_dict[i]['folate']['percentDailyValue'])
    caloriesFromFat_list.append(nutritions_dict[i]['caloriesFromFat']['percentDailyValue'])
    calcium_list.append(nutritions_dict[i]['calcium']['percentDailyValue'])
    fiber_list.append(nutritions_dict[i]['fiber']['percentDailyValue'])
    magnesium_list.append(nutritions_dict[i]['magnesium']['percentDailyValue'])
    iron_list.append(nutritions_dict[i]['iron']['percentDailyValue'])
    cholesterol_list.append(nutritions_dict[i]['cholesterol']['percentDailyValue'])
    protein_list.append(nutritions_dict[i]['protein']['percentDailyValue'])
    vitaminA_list.append(nutritions_dict[i]['vitaminA']['percentDailyValue'])
    potassium_list.append(nutritions_dict[i]['potassium']['percentDailyValue'])
    saturatedFat_list.append(nutritions_dict[i]['saturatedFat']['percentDailyValue'])
    vitaminC_list.append(nutritions_dict[i]['vitaminC']['percentDailyValue'])
    
nutritions_data = {'niacin': niacin_list, 'sugars': sugars_list, 'sodium': sodium_list, 
        'carbohydrates': carbohydrates_list, 'vitaminB6': vitaminB6_list, 'calories': calories_list,
        'thiamin': thiamin_list, 'fat': fat_list, 'folate': folate_list,
        'caloriesFromFat': caloriesFromFat_list, 'calcium': calcium_list, 'fiber': fiber_list, 
        'magnesium': magnesium_list, 'iron': iron_list, 'chalesterol': cholesterol_list,
        'protein': protein_list, 'vitaminA': vitaminA_list, 'potassium': potassium_list,
        'saturatedFat': saturatedFat_list, 'vitaminC': vitaminC_list}

nutritions_df = pd.DataFrame(nutritions_data)
nutritions_df.index = recipe['recipe_id']

cooking_directions_dict = []
for row in recipe['cooking_directions']:
    cooking_directions_dict.append(ast.literal_eval(row))
    
prep_list = []
cook_list = []
ready_in_list = []
directions_list = []
directions_in_list_form = 0

for i in range(len(cooking_directions_dict)):
    temp = list(cooking_directions_dict[i].values())[0].split('\n')
    try:
        if temp[0] == 'Prep':
            if temp[2] == 'Cook':
                prep_list.append(temp[1])
                cook_list.append(temp[3])
                ready_in_list.append(temp[5])
                directions_list.append(' '.join(temp[6:]))
            elif temp[2] == 'Ready In':
                prep_list.append(temp[1])
                cook_list.append('NaN')
                ready_in_list.append(temp[3])
                directions_list.append(' '.join(temp[4:]))
        elif temp[0] == 'Cook':
            prep_list.append('NaN')
            if temp[2] == 'Ready In':
                cook_list.append(temp[1])
                ready_in_list.append(temp[3])
                directions_list.append(' '.join(temp[4:]))
            else:
                cook_list.append(temp[1])
                ready_in_list.append('NaN')
                directions_list.append(' '.join(temp[2:]))
        elif temp[0] == 'Ready In':
            prep_list.append('NaN')
            cook_list.append('NaN')
            ready_in_list.append(temp[1])
            directions_list.append(' '.join(temp[2:]))
        else:
            prep_list.append('NaN')
            cook_list.append('NaN')
            ready_in_list.append('NaN')
            directions_list.append(' '.join(temp))
            directions_in_list_form += 1
    except:
        prep_list.append('NaN')
        cook_list.append('NaN')
        ready_in_list.append('NaN')
        directions_list.append('NaN')

cooking_directions_data = {'prep': prep_list, 'cook': cook_list, 'ready_in': ready_in_list, 'directions': directions_list}

cooking_directions_df = pd.DataFrame(cooking_directions_data)
cooking_directions_df.index = recipe['recipe_id']

# Separating the ingredients
ingredients_list = []
temp=[]
for row in range(len(recipe.ingredients)):
    temp = recipe.ingredients[row].split("^")
    ingredients_list.append(temp)
    temp = []

#  Cleaning the ingredients
ingredients_list_clean = []  
num_ingredients_list = []
for i in range(len(ingredients_list)):
    for j in range(len(ingredients_list[i])):
        if ":" not in ingredients_list[i][j]:
            temp.append(ingredients_list[i][j])
    ingredients_list_clean.append(temp)
    num_ingredients_list.append(len(temp))
    temp = []
    
ingredients_dict = {"ingredients":ingredients_list_clean,"number_ingredients":num_ingredients_list}
ingredients_df = pd.DataFrame(ingredients_dict)
ingredients_df.index= recipe["recipe_id"]

recipe_1 = recipe.drop(["ingredients","cooking_directions","image_url","cooking_directions","nutritions"],axis = 1)

recipe_clean = pd.merge(recipe_1, ingredients_df, on = "recipe_id")
recipe_clean = pd.merge(recipe_clean, cooking_directions_df, on = "recipe_id")
recipe_clean = pd.merge(recipe_clean, nutritions_df, on = "recipe_id")


recipe_clean = recipe_clean.replace('NaN', np.nan)
recipe_clean = recipe_clean.dropna()

def calc_time(row):
    breakdown = row.split(" ")
    if len(breakdown) == 4:
        h = breakdown[0]
        m = breakdown[2]
    elif len(breakdown) == 2 and "h" in row:
        h = breakdown[0]
        m = 0
    elif len(breakdown) == 2 and "m" in row:
        h = 0
        m = breakdown[0]
    elif len(breakdown) == 2 and "d" in row:
        h = 24*int(breakdown[0])
        m = 0
    elif len(breakdown) == 6:
        h = int(breakdown[2])+24*int(breakdown[0])
        m = breakdown[4]
    return int(m)+60*int(h)

recipe_clean["total_time"] = recipe_clean["ready_in"].apply(calc_time)

def get_n_steps(row):
    return len(row.split('.'))

recipe_clean["n_steps"] = recipe_clean["directions"].apply(get_n_steps)

display(recipe_clean.head())
```

Kaggle (Food.com; https://www.kaggle.com/shuyangli94/food-com-recipes-and-user-interactions?select=RAW_recipes.csv

```python
food = pd.read_csv(r'C:\Users\Nick\Desktop\data\RAW_recipes.csv')
food = food[["name", "minutes", "nutrition", "n_steps", "steps", "ingredients", "n_ingredients"]]
food['nutrition'] = food['nutrition'].apply(ast.literal_eval)
food['steps'] = food['steps'].apply(ast.literal_eval)
food['ingredients'] = food['ingredients'].apply(ast.literal_eval)

nutrition_df = pd.DataFrame(food['nutrition'].to_list(), columns=['calories','total_fat_pdv', "sugar_pdv", "sodium_pdv", "protein_pdv", "sat_fat_pdv", "carbs_pdv"])
food = food.merge(nutrition_df, left_index=True, right_index=True)[["name", "minutes", "n_steps", "steps", "ingredients", "n_ingredients", "calories", "carbs_pdv", "sodium_pdv", "protein_pdv"]]

dv_carbs = 275 # g
dv_protein = 100 # g
dv_sodium = 2300 # mg

food["carbs"] = food["carbs_pdv"]*dv_carbs*0.01
food["sodium"] = food["sodium_pdv"]*dv_sodium*0.01
food["protein"] = food["protein_pdv"]*dv_protein*0.01
food = food[["name", "minutes", "n_steps", "steps", "ingredients", "n_ingredients", "calories", "carbs", "sodium", "protein"]]

food.head()
```
